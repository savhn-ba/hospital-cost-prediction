```python
# read your data and store it into a dataframe
df = pd.read_excel("/content/drive/MyDrive/BUAN 302/Hospital Dataset.xlsx")
```

```python
display(df.head())
df.info()
```

```python
# select important features
features = ['AGE', 'GENDER', 'KEY COMPLAINTS -CODE', 'BODY WEIGHT', 'BODY HEIGHT', 'HR PULSE', 'RR', 'PAST MEDICAL HISTORY CODE', 'HB', 'MODE OF ARRIVAL', 'TYPE OF ADMSN']

# Clean up column names by stripping whitespace to prevent KeyErrors
df.columns = df.columns.str.strip()

x=df[features]
y=df['TOTAL COST TO HOSPITAL']
```

```python
#convert categorical variables into dummy variables
x=pd.get_dummies(x, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
print("Model Coefficients:")
for feature, coef in zip(x_train.columns, model.coef_):
    print(f"{feature}: {coef}")

print(f"\nModel Intercept: {model.intercept_}")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = model.predict(x_test)

print("MAE:", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
columns_to_drop = [
    'PAST MEDICAL HISTORY CODE_Hypertension1',
    'PAST MEDICAL HISTORY CODE_hypertension2',
    'PAST MEDICAL HISTORY CODE_other',
    'MODE OF ARRIVAL_TRANSFERRED',
    'MODE OF ARRIVAL_WALKED IN',
    'HB',

]

# Drop the specified columns from x
x_dropped = x.drop(columns=columns_to_drop, errors='ignore')

# Re-split the data
x_train_new, x_test_new, y_train, y_test = train_test_split(x_dropped, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize StandardScaler
scaler = StandardScaler()
```

```python
# Fit on training data and transform both training and test data
x_train_scaled = scaler.fit_transform(x_train_new)
x_test_scaled = scaler.transform(x_test_new)
```

```python
# Re-train the model with scaled data
model_scaled = LinearRegression()
model_scaled.fit(x_train_scaled, y_train)
```

```python
# Print new coefficients and intercept
print("Model Coefficients (after scaling features):")
# Note: The coefficients will correspond to the scaled features
for feature, coef in zip(x_train_new.columns, model_scaled.coef_):
    print(f"{feature}: {coef}")

print(f"\nModel Intercept (after scaling features): {model_scaled.intercept_}")
```

```python
# Re-evaluate the model with scaled data
y_pred_scaled = model_scaled.predict(x_test_scaled)
print("\n Model Evaluation (after scaling features) ")
print("MAE:", mean_absolute_error(y_test, y_pred_scaled))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_scaled)))
print("R2 Score:", r2_score(y_test, y_pred_scaled))
```

```python
complaint_columns = [
    'KEY COMPLAINTS -CODE_CAD-DVD',
    'KEY COMPLAINTS -CODE_CAD-SVD',
    'KEY COMPLAINTS -CODE_CAD-TVD',
    'KEY COMPLAINTS -CODE_CAD-VSD',
    'KEY COMPLAINTS -CODE_NONE',
    'KEY COMPLAINTS -CODE_OS-ASD',
    'KEY COMPLAINTS -CODE_PM-VSD',
    'KEY COMPLAINTS -CODE_RHD',
    'KEY COMPLAINTS -CODE_other- heart',
    'KEY COMPLAINTS -CODE_other- respiratory',
    'KEY COMPLAINTS -CODE_other-general',
    'KEY COMPLAINTS -CODE_other-nervous',
    'KEY COMPLAINTS -CODE_other-tertalogy'
]

# Drop 'GENDER_M' and create 'complaint frequency'
# Ensure x_dropped is fresh from the last 'x' before any previous complaint_columns modifications

additional_cols_to_drop = [
    'PAST MEDICAL HISTORY CODE_Hypertension1',
    'PAST MEDICAL HISTORY CODE_hypertension2',
    'PAST MEDICAL HISTORY CODE_other',
    'MODE OF ARRIVAL_TRANSFERRED',
    'MODE OF ARRIVAL_WALKED IN',
    'HB',
    'GENDER_M' # Adding GENDER_M here
]

# Re-create x_dropped from the original x, dropping all specified columns including GENDER_M
x_temp = x.drop(columns=additional_cols_to_drop, errors='ignore')

# Create the new 'complaint frequency' feature by summing the specified columns from x_temp
x_temp['complaint frequency'] = x_temp[complaint_columns].sum(axis=1)

# Drop the individual complaint columns now that the aggregated feature is created
x_final_fe = x_temp.drop(columns=complaint_columns, errors='ignore')

# Re-split the data with the new feature
x_train_fe, x_test_fe, y_train, y_test = train_test_split(x_final_fe, y, test_size=0.2, random_state=42)

# Re-apply scaling
scaler_fe = StandardScaler()
x_train_scaled_fe = scaler_fe.fit_transform(x_train_fe)
x_test_scaled_fe = scaler_fe.transform(x_test_fe)

# Re-train the model with the new feature and scaled data
model_fe = LinearRegression()
model_fe.fit(x_train_scaled_fe, y_train)

# Print new coefficients and intercept
print("\nModel Coefficients:")
for feature, coef in zip(x_train_fe.columns, model_fe.coef_):
    print(f"{feature}: {coef}")
print(f"\nModel Intercept: {model_fe.intercept_}")

# Re-evaluate the model with the new feature and scaled data
y_pred_fe = model_fe.predict(x_test_scaled_fe)
print("\n--- Model Evaluation ---")
print("MAE:", mean_absolute_error(y_test, y_pred_fe))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_fe)))
print("R2 Score:", r2_score(y_test, y_pred_fe))
```
